```python
# Create a groupby object the countries by Groups; That is grouging all US respondents together, 
#all Uk respondents together....etc.. Call it group_of_countries
group_of_countries = df.groupby('Country')
```

```python
# Use the groups_of_country to print this salary information
group_of_countries.sum(['Salary_in_USD'])['Salary_in_USD']
```

```python
# Use a for loop through the list ['Nepal', 'Mali', 'United States','United Kingdom','Canada','France']
# and use the printing style to print this
countries = ['Nepal', 'Mali', 'United States','United Kingdom','Canada','France']
for c in countries:
    print(c+"'s top 10 'SocialMedia")
    #print(group_of_countries.get_group(c)['SocialMedia'].value_counts())
    cv = group_of_countries.get_group(c)['SocialMedia'].value_counts() # value_counts
    #print(cv)
    f1_width = max([len(s) for s in cv.index])+1
    for i in range(len(cv)):
        print("%s\t%d" % ( cv.index[i].ljust(f1_width), cv[i]) )
    print("************************************\n")
```

```python
# Do the same thing to print this
countries = ['Nepal', 'Mali', 'United States','United Kingdom','Canada','France']
for c in countries:
    print(c+"'s top 10 'Salary_in_USD")
    cv =group_of_countries.get_group(c)['Salary_in_USD'].value_counts().sort_index(ascending=False).head(10)
    print(cv)
    print("_ _ _ _ _ _ _ _ _ _ _ _ _ _ _\n")
    
```

```python
# Do the same thing to print this
countries = ['Nepal', 'Mali', 'United States','United Kingdom','Canada','France']
for c in countries:
    summary = group_of_countries.get_group(c)['Salary_in_USD'].describe()
    print(c+'\n')
    print(" mean\t%f" % summary['mean'])
    print("max\t%f" % summary['max'])
    print("median\t%f" % summary['50%'])
   
```
